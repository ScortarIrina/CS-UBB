### Task

```python
import torch
import torch.nn as nn
import torch.optim as optim
from collections import OrderedDict


print("\n~~~~~~~~~~~~~~~~~~~~~~~~~~~~ MODEL 1 ~~~~~~~~~~~~~~~~~~~~~~~~~~~~\n")

# add 2 bits
data_in1 = torch.tensor([[0, 0], [0, 1], [1, 0], [1, 1]]).float()   
data_target1 = torch.tensor([[0, 0], [0, 1], [0, 1], [1, 0]]).float()

# define the model architecture
model1 = nn.Sequential(
    nn.Linear(2,4),  # hidden 1 net layer
    nn.ReLU(),       # hidden 1 activation function
    nn.Linear(4,8),  # hidden 2 net layer
    nn.Tanh(),       # hidden 2 activation function
    nn.Linear(8,2),  # output net layer
    nn.Sigmoid())    # output activation layer

print("\n1) MODEL:\n\n", model1, "\n")
print("\n2) WEIGHTS:\n\n", model1[0].weight, "\n")
print("\n3) BIAS\n\n:", model1[0].bias, "\n")

# binary cross entropy
criterion1 = nn.MSELoss()  
optimizer1 = optim.SGD(model1.parameters(), lr=0.2)

# train the model
epochs = 1000

for epoch in range(epochs):
    # forward pass
    outputs1 = model1(data_in1) 
    # compute loss
    loss = criterion1(outputs1, data_target1) 
    # backward pass and optimize
    optimizer1.zero_grad()
    loss.backward()
    optimizer1.step()
    # progress
    if (epoch + 1) % 100 == 0:
        print(f'epoch {epoch+1}/{epochs}, loss {loss.item():.4f}')
        
```
